```python
import pandas as pd
import matplotlib.pyplot as plt
import os 
```

```python
df =pd.read_csv("C:/Users/anush/world_population_data.csv")
```

```python
df.head()
```

```python
top_countries = df[['country', '2023 population']].sort_values(by='2023 population', ascending=False).head(10)
```

```python
plt.figure(figsize=(10, 6))
plt.barh(top_countries['country'], top_countries['2023 population'], color='skyblue')
plt.xlabel('Population in 2023')
plt.ylabel('Country')
plt.title('Top 10 Countries by Population in 2023')
plt.gca().invert_yaxis() 
plt.tight_layout()
plt.show()
```
